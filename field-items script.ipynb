```python
def field_items_scraper(url):
    
    names = []
    driver = webdriver.Chrome()
    driver.get(url)
    
    field_soup = driver.find_element_by_class_name("field-items")
    
    if url == "https://www.csun.edu/science-mathematics/biology/faculty-staff":
        p_tags = driver.find_elements_by_xpath("html/body/div[3]/div/div/div[3]/article/div/div/div/p")
        names = [tag.text for tag in p_tags]
        while "" in names:
            names.remove("")
        clean_names = []
        for name in names:
            colon_pos = name.find(":")
            comma_pos = name.find(",")
            left_paren_pos = name.find("(")
            if colon_pos != -1:
                name = name[:colon_pos]
            if comma_pos != -1:
                name = name[:comma_pos]
            if left_paren_pos != -1:
                name = name[:left_paren_pos]
            clean_names.append(name)
        names = clean_names
        names = names[2:101]
    
    elif url == 'https://www.csun.edu/humanities/american-indian-studies/faculty-staff' :
        raw_names = driver.find_elements_by_xpath('html/body/div[3]/div/div/div[3]/article/div/div/div/h2')
        raw_names = [name.text for name in raw_names]
        names = []
        for name in raw_names:
            comma_pos = name.find(",")
            names.append(name[:comma_pos])


    # this one is still tough
    elif url == 'https://www.csun.edu/health-human-development/health-sciences/lecturers':
        raw_names = driver.find_elements_by_xpath("html/body/div[3]/div[2]/div/div[3]/article/div/div/div/p")
        raw_names2 = driver.find_elements_by_xpath("html/body/div[3]/div[2]/div/div[3]/article/div/div/div/blockquote/p[2]")
        
        raw_names = [web_obj.find_element_by_tag_name("a") for web_obj in raw_names]
        names = [name.text for name in raw_names]
        
        #raw_names2 = [web_obj.find_element_by_tag_name("p") for web_obj in raw_names2]
        names2 = [name.text for name in raw_names2]
        
        names += names2
        
        #raw_names = [name.get_attribute("innerHTML") for name in raw_names]
        #for name in raw_names:
        #    print(name)
        
        
    # if it has a table
    elif field_soup.find_elements_by_tag_name("tbody"):
        print("table script ran")
        field_soup = field_soup.find_element_by_tag_name("tbody")
        if field_soup.find_elements_by_xpath("//tr/th"):
            print("a")
            web_objects = field_soup.find_elements_by_xpath("//tr/th")
            names = [web_object.text for web_object in web_objects]
        elif field_soup.find_elements_by_xpath("//tr/td/h3/img"):
            print("b")
            web_objects = field_soup.find_elements_by_xpath("//h3/img")
            names = [web_object.get_attribute("alt") for web_object in web_objects]
        elif field_soup.find_elements_by_xpath("//tr/td/h2"):
            print("c")
            web_objects = field_soup.find_elements_by_xpath("//tr/td/h2")
            names = [web_object.text for web_object in web_objects]
        elif field_soup.find_elements_by_xpath('//tr/td/ul/li[1]/h2'):
            print("d")
            web_objects = field_soup.find_elements_by_xpath('//tr/td/ul/li[1]/h2')
            names = [web_object.text for web_object in web_objects]
        else:
            print("e")
            field_soup = field_soup.find_elements_by_tag_name("tr")
            web_objects = []
            for row in field_soup:
                web_objects.append(row.find_element_by_tag_name("td"))
            field_soup = web_objects
            fields = [field for field in field_soup]
            names = [field.text for field in fields]
        print(names)
    
    # this really needs to be cleaned up
    elif field_soup.find_elements_by_xpath("//div/blockquote/p/strong"):
        print("blockquote script 1 ran")
        first_name_objects = field_soup.find_elements_by_xpath("//p/strong/a")
        last_name_objects = field_soup.find_elements_by_xpath("//div/blockquote/p/a/strong")
        name_objects = field_soup.find_elements_by_xpath("//div/p/a/strong")
        first_names = [web_object.text for web_object in first_name_objects]
        last_names = [web_object.text for web_object in last_name_objects]
        more_names = [web_object.text for web_object in name_objects]

        print(first_names)
        print(last_names)
        print(more_names)
        
        # still need to combine all these into names
        
    elif field_soup.find_elements_by_xpath("//div/blockquote/p/a"):
        print("blockquote script 2 ran")
        name_objects = field_soup.find_elements_by_xpath("//div/blockquote/p/a")
        names = [web_object.text for web_object in name_objects]
        
    elif field_soup.find_elements_by_xpath("//div/blockquote/p"):
        print("blockquote script 3 ran")
        name_objects = field_soup.find_elements_by_xpath("//div/blockquote/p/a")
        names = [web_object.text for web_object in name_objects]
        
    # have to make this a bit less general
    elif field_soup.find_elements_by_tag_name("li"):
        print("list script ran")
        field_soup = field_soup.find_elements_by_tag_name("li")
        fields = [field for field in field_soup]
        names = [field.text for field in fields]
        print(names)
        
    # too general
    elif field_soup.find_elements_by_xpath("//div/h2"):
        print("h2 script outside table ran")
        web_objects = field_soup.find_elements_by_xpath("//div/h2")
        names = [web_object.text for web_object in web_objects]
        names = names[4]
        
    elif field_soup.find_elements_by_xpath("//a"):
        print("a script ran")
        a_soup = field_soup.find_elements_by_tag_name("//a")
        fields = [field for field in a_soup]
        names = [field.text for field in fields]
    
    else:
        print("Didn't Scrape")
    
    driver.close()
    
    return names
```
