```python
def field_items_scraper(url):
    
    names = []
    driver = webdriver.Chrome()
    driver.get(url)
    
    field_soup = driver.find_element_by_class_name("field-items")
    
    if url == "https://www.csun.edu/science-mathematics/biology/faculty-staff":
        print("First Custom Ran")
        p_tags = driver.find_elements_by_xpath("html/body/div[3]/div/div/div[3]/article/div/div/div/p")
        names = [tag.text for tag in p_tags]
        while "" in names:
            names.remove("")
        clean_names = []
        for name in names:
            colon_pos = name.find(":")
            comma_pos = name.find(",")
            left_paren_pos = name.find("(")
            if colon_pos != -1:
                name = name[:colon_pos]
            if comma_pos != -1:
                name = name[:comma_pos]
            if left_paren_pos != -1:
                name = name[:left_paren_pos]
            clean_names.append(name)
        names = clean_names
        names = names[2:101]
    
    elif url == 'https://www.csun.edu/humanities/american-indian-studies/faculty-staff' :
        print("Second Custom Ran")
        raw_names = driver.find_elements_by_xpath('html/body/div[3]/div/div/div[3]/article/div/div/div/h2')
        raw_names = [name.text for name in raw_names]
        names = []
        for name in raw_names:
            comma_pos = name.find(",")
            names.append(name[:comma_pos])


    # should be done
    elif url == 'https://www.csun.edu/health-human-development/health-sciences/lecturers':
        print("Third Custom Ran")
        raw_names = driver.find_elements_by_xpath("html/body/div[3]/div[2]/div/div[3]/article/div/div/div/p")
        raw_names = [web_obj.find_element_by_tag_name("a") for web_obj in raw_names]
        names1 = [name.text for name in raw_names]

        raw_names2 = driver.find_elements_by_xpath("html/body/div[3]/div[2]/div/div[3]/article/div/div/div/blockquote")
        names2 = [name.text for name in raw_names2]
        temp = names2
        temp = [name.split("\n\n") for name in temp]
        for name in temp:
            if len(name) == 1:
                temp.remove(name)

        for name in temp:
            if len(name) == 1:
                temp.remove(name)

        names = []
        for name_list in temp:
            for name in name_list:
                email_pos = name.find("Email")
                if email_pos != -1:
                    names.append(name[:email_pos])


        temp = names
        new_names = [name.split("\n") for name in temp]

        for name_list in new_names:
            for name in name_list:
                if len(name) == 0:
                    name_list.remove(name)

        for name_list in new_names:
            for name in name_list:
                if len(name) == 1:
                    name_list.remove(name)
                    
        names = []
        for name in new_names:
            names.append(name[0])
        names += names1

        clean_names = []
        for name in names:
            clean_names.append(name.strip())
        names = clean_names
        
        
    # if it has a table
    elif field_soup.find_elements_by_tag_name("tbody"):
        print("table script ran")
        field_soup = field_soup.find_element_by_tag_name("tbody")
        if field_soup.find_elements_by_xpath("//tr/th"):
            print("a")
            web_objects = field_soup.find_elements_by_xpath("//tr/th")
            names = [web_object.text for web_object in web_objects]
        elif field_soup.find_elements_by_xpath("//tr/td/h3/img"):
            print("b")
            web_objects = field_soup.find_elements_by_xpath("//h3/img")
            names = [web_object.get_attribute("alt") for web_object in web_objects]
        elif field_soup.find_elements_by_xpath("//tr/td/h2"):
            print("c")
            web_objects = field_soup.find_elements_by_xpath("//tr/td/h2")
            names = [web_object.text for web_object in web_objects]
        elif field_soup.find_elements_by_xpath('//tr/td/ul/li[1]/h2'):
            print("d")
            web_objects = field_soup.find_elements_by_xpath('//tr/td/ul/li[1]/h2')
            names = [web_object.text for web_object in web_objects]
        else:
            print("e")
            field_soup = field_soup.find_elements_by_tag_name("tr")
            web_objects = []
            for row in field_soup:
                web_objects.append(row.find_element_by_tag_name("td"))
            field_soup = web_objects
            fields = [field for field in field_soup]
            names = [field.text for field in fields]
        print(names)
    
    # this really needs to be cleaned up
    elif field_soup.find_elements_by_xpath("//div/blockquote/p/strong") or url == 'https://www.csun.edu/health-human-development/health-sciences/adjunct-faculty':
        print("blockquote script 1 ran")
        
        if url == 'https://www.csun.edu/health-human-development/health-sciences/adjunct-faculty':
            name_objects = field_soup.find_elements_by_xpath("//div/p")
            name_objects += field_soup.find_elements_by_xpath("//div/blockquote/p")
            names = [web_object.text for web_object in name_objects]
            for name in names:
                if len(name) == 0:
                    names.remove(name)
            for name in names:
                if len(name) == 1:
                    names.remove(name)
            temp = []
            for name in names:
                newline_pos = name.find("\n")
                if newline_pos != -1:
                    temp.append(name[:newline_pos])
                
            names = [name.strip() for name in temp]

        else:
            first_name_objects = field_soup.find_elements_by_xpath("//p/strong/a")
            last_name_objects = field_soup.find_elements_by_xpath("//div/blockquote/p/a/strong")
            name_objects = field_soup.find_elements_by_xpath("//div/p/a/strong")
            first_names = [web_object.text for web_object in first_name_objects]
            last_names = [web_object.text for web_object in last_name_objects]
            more_names = [web_object.text for web_object in name_objects]
            names.append(first_names[0] + last_names[0])
            first_names.remove(first_names[0])
            last_names.remove(last_names[0])
            names.append(last_names[0])
            names += more_names + first_names
        
        # still need to combine all these into names
        
    elif field_soup.find_elements_by_xpath("//div/blockquote/p/a"):
        print("blockquote script 2 ran")
        name_objects = field_soup.find_elements_by_xpath("//div/blockquote/p/a")
        names = [web_object.text for web_object in name_objects]
        
    elif field_soup.find_elements_by_xpath("//div/blockquote/p"):
        print("blockquote script 3 ran")
        name_objects = field_soup.find_elements_by_xpath("//div/blockquote/p/a")
        names = [web_object.text for web_object in name_objects]
        
    # have to make this a bit less general
    elif field_soup.find_elements_by_tag_name("li"):
        print("list script ran")
        field_soup = field_soup.find_elements_by_tag_name("li")
        fields = [field for field in field_soup]
        names = [field.text for field in fields]
        print(names)
        
    # too general
    elif field_soup.find_elements_by_xpath("//div/h2"):
        print("h2 script outside table ran")
        web_objects = field_soup.find_elements_by_xpath("//div/h2")
        names = [web_object.text for web_object in web_objects]
        names = names[4]
        
    elif field_soup.find_elements_by_xpath("//a"):
        print("a script ran")
        a_soup = field_soup.find_elements_by_tag_name("//a")
        fields = [field for field in a_soup]
        names = [field.text for field in fields]
    
    else:
        print("Didn't Scrape")
    
    driver.close()
    
    return names
```

```python
counter = 0
all_names = []
for url in urls:
    print("URL ", counter)
    all_names.append(field_items_scraper(url))
    counter += 1
```

```python
# cleaning names 0
temp = []
for name in all_names[0]:
    temp.append(name.replace("Professor"," "))
names = []
for name in temp:
    comma_pos = name.find(",")
    left_paren = name.find("(")
    if comma_pos != -1:
        name = name[:comma_pos]
    if left_paren != -1:
        name = name[:left_paren]
    names.append(name.strip())
all_names[0] = names
```

```python
# cleaning names 1
temp = []
for name in all_names[1]:
    temp.append(name.replace("Professor", " "))
names = []
for name in temp:
    names.append(name.strip())
all_names[1] = names
```

```python
# cleaning names 2
temp = []
for name in all_names[2]:
    temp.append(name.replace("Professor", " "))
names = []
for name in temp:
    names.append(name.strip())
all_names[2] = names
```

```python
# cleaning names 3
temp = []
for name in all_names[3]:
    temp.append(name.split(","))
names = []
for name in temp:
    names.append(name[0] + name[1])
    
all_names[3] = names
```

```python
# cleaning 4
temp = []
for name in all_names[4]:
    temp.append(name.replace("Dr."," "))
names = []
for name in temp:
    comma_pos = name.find(",")
    if comma_pos != -1:
        names.append(name[:comma_pos])
    else:
        names.append(name)
        
names = [name.strip() for name in names]
        
all_names[4] = names
```

```python
# cleaning 11

# this is only a single string
names = []
comma_pos = all_names[11].find(",")
all_names[11] = all_names[11][:comma_pos]
```

```python
# cleaning 12
for i in range(3):
    all_names[12].remove(all_names[12][0])
names = []
for name in all_names[12]:
    newline = name.find("\n")
    if name.find("\n") != -1:
        names.append(name[:newline])
    else:
        names.append(name)
all_names[12] = names
```

```python
# cleaning 13

names = []
for name in all_names[13]:
    newline = name.find("\n")
    if name.find("\n") != -1:
        names.append(name[:newline])
    else:
        names.append(name)

temp = []
for name in names:
    comma = name.find(",")
    if name.find("\n") != -1:
        temp.append(name[:comma])
    else:
        temp.append(name)

all_names[13] = temp
```

```python
# cleaning 24

names = []
for name in all_names[24]:
    newline = name.find("\n")
    if name.find("\n") != -1:
        names.append(name[:newline])
    else:
        names.append(name)

temp = []
for name in names:
    comma = name.find(",")
    if name.find("\n") != -1:
        temp.append(name[:comma])
    else:
        temp.append(name)

all_names[24] = temp
```
