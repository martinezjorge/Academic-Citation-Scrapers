```python
def journal_cleaner(journal):
    
    if ':' in journal and '(' in journal:
        thing = journal.split('(')
        journal_title= thing[0][:-2]
        pages = thing[1].split(':')
        issue = pages[0].replace(')'," ").strip()
        new_pages = pages[-1]
        thing2 = thing[0].split()
        volume = thing2[-1]
        data = {'journal':journal_title.strip(),
                'pages': new_pages,
                'volume' : volume
               }

    elif '(' in journal:
        temp = journal.split('(')
        #'Linear Algebra and its Applications 412 (s 2–3)' 
        
        journal_volume = temp[0].split()
        
        volume = journal_volume[-1]
        
        journal_title = ""
        for i,j in enumerate(journal_volume):
            if i < len(journal_volume) -1:
                journal_title += j + " "
            else:
                journal_title = journal_title[:-1]
                
        issue = temp[1].replace(")", " ").strip()
        
        data = {'journal':journal_title.strip(),
                'volume': volume,
                'issue':issue
               }
        
    elif ':' in journal:
        #  Lecture Notes in Computer Science 1671:168-176
        temp = journal.split(":")
        
        pages = temp[1]
        
        journal_volume = temp[0].split()
        
        journal_title = ""
        for i,j in enumerate(journal_volume):
            if i < len(journal_volume) -1:
                journal_title += j + " "
            else:
                journal_title = journal_title[:-1]
                
        volume = journal_volume[-1]
        
        data = {'journal':journal_title.strip(),
                'volume':volume,
                'pages':pages}
        
    else:
        journal_title = journal
        data = {'journal':journal_title.strip()}
        # Journal of Biomechanics 40
        
        # Human Performance
        
        
        #data = ("This case was not accounted for")

    return data
```

```python
all_faculty_names = all_faculty.faculty.tolist()
email_faculty = emails.faculty.tolist()
emails = emails.email.tolist()
len(all_faculty_names)
```

```python
email_faculty
```

```python
def get_matches(query,choices,limit=4):
    results = process.extract(query,[choices],limit=limit)
    return results
```

```python
indeces = []

for i, name in enumerate(all_faculty_names):
    
    for j, name2 in enumerate(email_faculty):
        
        matches = get_matches(name,name2)
        
        if matches[0][1] > 87:
            
            indeces.append([i,name,j,name2])
            
            break
        
        
```

```python
indeces
count = 0
for item in indeces:
    if item[1].strip() != item[3].strip():
        count +=1
        print(item)
print(count)
```

```python
with open("matches_third_try.txt", 'w') as file:
    for item in indeces:
        record = ""
        for p in item:
            record += str(p) + ","
        record = record[:-1]
        file.write(record + '\n') 
```
