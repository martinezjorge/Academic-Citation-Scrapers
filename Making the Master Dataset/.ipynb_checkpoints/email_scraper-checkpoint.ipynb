```python
def new_faculty_scraper():
    prof_name = []
    prof_email = []
    start_page = "https://academics.csun.edu/faculty/search"
    driver = webdriver.Chrome()
    driver.get(start_page)
    next_button_exists = False
    if driver.find_elements_by_xpath('//*[@id="main-content"]/div[2]/div[2]/div/ul/li[13]/a'):
        next_button = driver.find_element_by_xpath('//*[@id="main-content"]/div[2]/div[2]/div/ul/li[13]/a')
        next_button_exists = True
        names_list = open("names_with_emails.csv", "a")
        more_results = True
        while more_results:
            professor_names = driver.find_elements_by_xpath('//*[@id="main-content"]/div[2]/div[1]/div/div/div/div/h3/a')
            for name in professor_names:
                print(name.get_attribute('innerHTML'))
                prof_name.append(name.get_attribute('innerHTML'))
                print(name.get_attribute('href'))
                prof_email.append(name.get_attribute('href'))
                        
            if not driver.find_elements_by_xpath('//*[@id="main-content"]//li[14][@class="disabled"]'):
                #Click the next button and reselect the next button
                if next_button_exists:
                    next_button =  driver.find_element_by_xpath('//a[@rel="next"]')
                    webdriver.ActionChains(driver).move_to_element(next_button).click(next_button).perform()
                if driver.find_elements_by_xpath('//a[@rel="next"]'):
                    next_button =  driver.find_element_by_xpath('//a[@rel="next"]')
                else:
                    more_results = False
        return prof_name, prof_email
                                
        driver.close()    
            
                
    
            
```

```python
names_emails = new_faculty_scraper()
```

```python
def newer_faculty_scraper():
    link_list = []
    start_page = "https://academics.csun.edu/faculty/search"
    driver = webdriver.Chrome()
    driver.get(start_page)
    next_button_exists = False
    if driver.find_elements_by_xpath('//*[@id="main-content"]/div[2]/div[2]/div/ul/li[13]/a'):
        next_button = driver.find_element_by_xpath('//*[@id="main-content"]/div[2]/div[2]/div/ul/li[13]/a')
        next_button_exists = True
        names_list = open("new_names_with_emails.csv", "a")
        more_results = True
        while more_results:
            professor_names = driver.find_elements_by_xpath('//*[@id="main-content"]/div[2]/div[1]/div/div/div/div/h3/a')
            for name in professor_names:
                link_list.append(name.get_attribute('href'))                        
            if not driver.find_elements_by_xpath('//*[@id="main-content"]//li[14][@class="disabled"]'):
                #Click the next button and reselect the next button
                if next_button_exists:
                    next_button =  driver.find_element_by_xpath('//a[@rel="next"]')
                    webdriver.ActionChains(driver).move_to_element(next_button).click(next_button).perform()
                if driver.find_elements_by_xpath('//a[@rel="next"]'):
                    next_button =  driver.find_element_by_xpath('//a[@rel="next"]')
                else:
                    more_results = False
        return link_list
                                
        driver.close()  
            
```
