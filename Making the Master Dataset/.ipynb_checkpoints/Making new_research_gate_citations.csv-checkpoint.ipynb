```python
import pandas as pd
import numpy as np
```

```python
sandbox_all_faculty = pd.read_csv('sandbox_all_faculty.csv')
all_faculty = pd.read_csv('all_faculty.csv')
final = pd.read_csv('final_matches.csv')
nerces = pd.read_excel('facult_query_request_2.xlsx')
rg = pd.read_csv('research_gate_citations.csv')
```

```python
sandbox_all_faculty = sandbox_all_faculty.drop(['sandbox'], axis=1)
```

```python
sandbox_all_faculty.head()
```

```python
final = final.drop(['email'],axis=1)
```

```python
final.head()
```

## Outer join

```python
new_final_matches_outer = pd.merge(final,sandbox_all_faculty[['faculty','email']],on='faculty',how='outer')
```

```python
new_final_matches_outer.shape[0]
```

```python
new_final_matches_outer.head()
```

```python
new_final_matches_outer.to_csv('new_final_matches_outer.csv',index=False)
```

```python
new_final_matches_outer = pd.read_csv('new_final_matches_outer.csv')
```

```python
new_final_matches_outer.head()
```

```python
new_rg_2 = pd.merge(rg,new_final_matches_outer[new_final_matches_outer.columns],on='faculty',how='inner')
```

```python
new_final_matches.to_csv('new_final_matches.csv',index=False)
```
