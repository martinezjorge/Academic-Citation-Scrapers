```python
faculty = data.faculty.tolist()
citation = data.citation.tolist()
title = data.title.tolist()
paper_type = data.type.tolist()
abstract = data.abstract.tolist()
book = data.book.tolist()
journal = data.journal.tolist()
date = data.date.tolist()
publisher = data.publisher.tolist()
edition = data.edition.tolist()
series = data.series.tolist()
volume = data.volume.tolist()
issue = data.issue.tolist()
chapter = data.chapter.tolist()
pages = data.pages.tolist()
doi = data.doi.tolist()
issn = data.issn.tolist()
isbn = data.isbn.tolist()
handle = data.handle.tolist()
url = data.url.tolist()
email = data.email.tolist()
sandbox=data.sandbox.tolist()
```

```python
api_format = []
for i in range(len(title)):
    dict_template = {
      "status": "",
      "success": "",
      "api": "citations",
      "version": "1.0",
      "collection": "",
      "count": "",
      "article": {
        "citation_id": "",
        "formatted": citation[i],
        "note": "",
        "membership": {
          "type": "",
          "members": [
            {
              "user_id": "",
              "orcid": "",
              "email": email[i],
              "display_name": "",
              "first_name": "",
              "middle_name": "",
              "last_name": "",
              "profile": sandbox[i],
              "precedence": "",
              "role": "author"
            }
          ]
        },
        "published": {
          "how": "",
          "date": date[i]
        },
        "is_published": "",
        "type": paper_type[i],
        "metadata": {
          "title": title[i],
          "abstract": abstract[i],
          "book_title": book[i],
          "journal": journal[i]
        },
        "collection": "",
        "document": "",
        "publisher": publisher[i],
      }
    }
    api_format.append(dict_template)
```

```python
with open('master_dataset.json','w') as file:
    for item in api_format:
        jsoned_dict = json.dumps(item)
        file.write(jsoned_dict + "\n")
```
