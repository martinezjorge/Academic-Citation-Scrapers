```python
import pandas as pd
from selenium import webdriver
```

```python
def catalog_content_script(url):
    driver = webdriver.Chrome()
    driver.get(url)
    content_element = driver.find_element_by_id("catalog_content")
    content_html = content_element.get_attribute("innerHTML")
    names = content_element.find_elements_by_class_name("jewel-display-name")
    names = [name.text for name in names]
    print(names)
    driver.close()
    return names
```

```python
college_data = pd.read_csv("department_urls.csv")
college_data
```

```python
catalog_content = college_data[college_data.id == 'catalog_content']
```

```python
catalog_content
```

```python
catalog_urls = catalog_content.url.tolist()
```

```python
for url in catalog_urls:
    catalog_content_script(url)
```
