```python
import pandas as pd
from selenium import webdriver
```

```python
def catalog_content_script(url):
    driver = webdriver.Chrome()
    driver.get(url)
    content_element = driver.find_element_by_id("catalog_content")
    content_html = content_element.get_attribute("innerHTML")
    names = content_element.find_elements_by_class_name("jewel-display-name")
    names = [name.text for name in names]
    print(names)
    driver.close()
    return names
```

```python
college_data = pd.read_csv("department_urls.csv")
college_data
```

```python
catalog_content = college_data[college_data.id == 'catalog_content']
```

```python
catalog_content
```

```python
catalog_urls = catalog_content.url.tolist()
for url in catalog_urls:
    print(url)
```

```python
f = open('catalog_content.csv', "a")

for i in range(catalog_content.shape[0]):
    
    catalog_content.iloc[i]
    names = catalog_content_script(catalog_content.iloc[i].url)
    
    if catalog_content.iloc[i].url == "https://www.csun.edu/health-human-development/environmental-occupational-health/eoh-faculty-and-part-time-lecturer":
        new_names = []
        for name in names:
            comma_pos = name.find(",")
            if comma_pos != -1:
                new_names.append(name[:comma_pos])
        names = new_names
    else:
        new_names = []
        for name in names:
            comma_pos = name.find(",")
            phd_pos = name.find("Ph.D.")
            if comma_pos != -1:
                name = name.replace(",", "")
            if phd_pos != -1:
                name = name.replace("Ph.D.", "")
            new_names.append(name.strip())
        names = new_names
        
    for name in names:
        record = name + ", " + catalog_content.iloc[0].College + ", " + catalog_content.iloc[0].Department + ", " + catalog_content.iloc[0].url + "\n" 
        f.write(record)
        
f.close()
```

```python
for url in catalog_urls:
    catalog_content_script(url)
```
