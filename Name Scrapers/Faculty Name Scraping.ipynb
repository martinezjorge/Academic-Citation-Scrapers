```python
def get_faculty_names(department):
    driver = webdriver.Chrome()
    driver.get(department)
    content_element = driver.find_element_by_id("catalog_content")
    content_html = content_element.get_attribute("innerHTML")
    faculty_soup = soup(content_html, "html.parser")
    names = faculty_soup.find_all("h3", {"class":"jewel-display-name"})
    faculty_names = [name.text for name in names]
    len(faculty_names)
    print(faculty_names)
    driver.close()
    return faculty_names
```

```python
for url in catalog_content_urls:
    get_faculty_names(url)
```
