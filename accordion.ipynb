```python
import pandas as pd
from selenium import webdriver
import time
```

```python
college_data = pd.read_csv("department_urls.csv")
college_data_accordion = college_data[college_data.id == 'accordion']
college_data_accordion_urls = college_data_accordion.url.tolist()
```

```python
counter = 0
for url in college_data_accordion_urls:
    print(counter,":",url)
    counter += 1
```

```python
def accordion_name_scrape(url): 

    driver = webdriver.Chrome(executable_path=r'C:\chromedriver.exe')
    driver.get(url)
    
    names = []      
    accordion_elements = driver.find_elements_by_xpath('//h2[@class="field field-name-field-title-text field-type-text field-label-hidden ui-accordion-header ui-helper-reset ui-state-default ui-corner-all"]')
   
    
    if url == "https://www.csun.edu/health-human-development/family-consumer-sciences/faculty-staff" or url == "https://www.csun.edu/finance/faculty-directory" or url == "https://www.csun.edu/management/faculty-directory" or url == "https://www.csun.edu/marketing/faculty-directory" or url == "https://www.csun.edu/som/faculty-staff-directory" or url == "https://www.csun.edu/mike-curb-arts-media-communication/art/faculty-staff": 
        tab_buttons = driver.find_elements_by_xpath('//*[@id="accordion"]/h2/span')
        for tab_button in tab_buttons:
            webdriver.ActionChains(driver).move_to_element(tab_button).click(tab_button).perform()
            time.sleep(1)
            
            list_elements = driver.find_elements_by_xpath('//*[@id="accordion"]/div/ul/li')
            for list_element in list_elements:
                names.append(list_element.text)
        while '' in names:
            names.remove('')

               
    elif url == "https://www.csun.edu/humanities/modern-classical-languages-literatures/faculty-staff":
        tab_buttons = driver.find_elements_by_xpath('//*[@id="accordion"]/h2/span')
        for tab_button in tab_buttons:
            webdriver.ActionChains(driver).move_to_element(tab_button).click(tab_button).perform()
            time.sleep(1)
            
            strong_elements = driver.find_elements_by_tag_name('strong')  
            for strong_element in strong_elements:
                names.append(strong_element.text)
        while '' in names:
            names.remove('')
        
    elif url ==  "https://www.csun.edu/acctis/faculty-and-staff-directory":
        tab_buttons = driver.find_elements_by_xpath('//*[@id="accordion"]/h2/span')
        for tab_button in tab_buttons:
            webdriver.ActionChains(driver).move_to_element(tab_button).click(tab_button).perform()
            time.sleep(1)
            
            h4_elements = driver.find_elements_by_xpath('//*[@id="accordion"]/div/ul/li/h4')
            for h4_element in h4_elements:
                names.append(h4_element.text)
        while '' in names:
            names.remove('')
            

    else:
        for accordion_element in accordion_elements:
            names.append(accordion_element.text)
            
    return names
        
        
    driver.close()
```

```python
temp = college_data_accordion_urls[0]
accordion_name_scrape(temp)
```

```python
scrape_array = []
for url in college_data_accordion_urls:
     scrape_array.append(accordion_name_scrape(url))
scrape_array
```
